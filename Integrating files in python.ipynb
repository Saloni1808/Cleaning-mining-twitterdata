```python
f = open('/Users/saurabhprasad/Downloads/Filtered_result_completeBrussels.csv')
raw = f.read()

tokens = nltk.word_tokenize(raw)

#Create your bigrams
bgs = nltk.quadgrams(tokens)

#compute frequency distribution for all the bigrams in the text
newdic = {}
fdist = nltk.FreqDist(bgs)
for k,v in fdist.items():
    newdic[k]=v
```
